```python
df.columns = (df.columns.str.lower()
                .str.replace(" ", "_")
                .str.replace("\xa0", "_")
                .str.replace(r"[^\w]", ""))
df[["actual_gross", "average_gross", "adjusted_gross_(in_2022_dollars)"]] = df[["actual_gross", "average_gross", "adjusted_gross_(in_2022_dollars)"]].replace(r"[^0-9]", "", regex=True)
df["tour_title"] = df["tour_title"].str.replace(r"\[.*?\]","", regex = True).str.replace(r"[^\w]", "", regex = True).str.strip()
```

```python
df.info()
```

```python
df.drop(columns=["peak", "all_time_peak"] ,inplace=True)
```

```python
df[['ref.']].T
#useless column
df.drop(columns=['ref.'], inplace= True)
```

```python
df.head(20)
```

```python

newyears = df["year(s)"].str.split("–", expand= True)

```

```python
newyears
```

```python
df = df.rename(columns={"year(s)" : "start_year"})
df["start_year"] = newyears[0]
df["end_year"] = newyears[1]
```

```python
col_end = df.pop("end_year")
df.insert(6, "end_year", col_end)
df['end_year'] = df['end_year'].fillna(df['start_year'])
```

```python
df.to_csv("cleaned", index=False)
```
